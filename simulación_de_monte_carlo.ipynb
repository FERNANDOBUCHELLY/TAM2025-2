```python
import numpy as np
import matplotlib.pyplot as plt

M = 1000 # numero de simulaciones de monte carlo
N = 100
A = 5
A1 = np.zeros((M,1))
A2 = np.zeros((M,1))

for m in range(M): # una iteracion de monte carlo
  print(f'simulacion {m+1}/{M} de Monte carlo...')
  w = np.random.randn(N) # me entrega un nuevo set de datos
  x = A + w
  A1[m] = np.mean(x) # estimador 1--> promedio iid verosimitud por gaussiana
  A2[m] = x[0] #estimador 2 --> promedio = muestra 1

plt.hist(A1, bins=20, label='Est_1', alpha = 0.5)
plt.hist(A2, bins=20, label='Est_2', alpha = 0.5)
plt.legend()
plt.show()
plt.plot(w, label='w')
plt.plot(x, label='x')
plt.legend()
plt.show()

```

```python
mw = np.mean(w)
mx = np.mean(x)
vw = np.var(w)
vx = np.var(x)
print(mw, mx)
print(vw,vx)
```
